```python

#defining a function
def append(s, val):
  for i in range(len(s)):
    if val < s[-1]:
      s.pop()
  s.append(val)
  return

def pop(s):
  s.pop()
  return

def __len__(s):
  return len(s)



#taking inputs from users
q = int(input("No. of queries: "))
s=[]
for i in range(q):
  a = input()
  b = a.split(" ")
  if b[0] == "size":
    print (__len__(s))
  elif b[0] == "pop":
    pop(s)
  elif b[0] =="append":
    append(s, int(b[1]))




```
