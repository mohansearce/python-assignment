```python
#define even funtion
def even (no_1, no_2):
  s=[]
  if no_1%2 == 0:
    for i in range (no_2):
      s.append(str(no_1+2*i))
  else:
    for i in range (no_2):
      s.append(str(no_1+1+2*i))


  x=" ".join(s)
  print(x)
  return

#take inputs from user
i = input("enter the two values: ")
j = i.split()
start = int(j[0])
n = int(j[1])

even (start, n)
```
